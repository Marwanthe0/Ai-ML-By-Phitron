```python
from abc import ABC,abstractmethod
class abstr(ABC):
    def __init__(self):
        pass
    @abstractmethod
    def show_info():
        pass

class Vehicle(abstr):
    def __init__(self,model,speed = 100):
        self.__model = model
        self.__speed = speed
    def classify(self):
        if(self.__speed > 100):
            return "High-speed Vehicle"
        else:
            return "Normal Vehicle"
    def show_info(self):
        print(f"Type: {self.__class__.__name__}, Name: {self.__model},Speed: {self.__speed}km/h, Predicted: {self.classify()}")
    def added_func(self):
        print(f"{self.__class__.__name__} Added: {self.__model}")

class Car(Vehicle):
    def __init__(self,model,speed = 100):
        super().__init__(model,speed)
class Bike(Vehicle):
    def __init__(self,model,speed = 100):
        super().__init__(model,speed)
v1 = Vehicle("Mercedez Benz-x",220)
c1 = Car("Toyota Premio 2020",150)
b1 = Bike("R15 v4",250)
# print(v1.classify())
# print(c1.classify())
# print(b1.classify())

n = int(input())
for i in range(n):
    a,b,c = tuple(map(str,input().split()))
    c = int(c)
    if(a.casefold() == "car"):
        c1 = Car(b,c)
        c1.added_func()
        c1.show_info()
    elif(a.casefold() == "bike"):
        b1 = Bike(b,c)
        b1.added_func()
        b1.show_info()
    else:
        print("Enter a valid Vehicle type")
        i -= 1
```
