Practice_Problem_01

```python
# Importing pandas as csv file
import pandas as pd

df = pd.read_csv("Practice_data.csv")
df
```

```python
# Creating new discounted column with 10% discounted price
df["Discounted_price"] = df["Price"] - (df["Price"] * (0.1))
```

```python
# Saving the new dataframe as discounted_products.csv
df.to_csv("discounted_products.csv", index=False)
```

Problem_02

```python
# From EmployeeID Column: Counting Duplicate ids
duplicate_id = df["EmployeeID"].duplicated().sum()
print("Number of Dupicate ids: ", duplicate_id)
```

```python
# Removing Duplicate ids and displaying cleaned DataFrame
df.drop_duplicates(subset="EmployeeID",inplace=True)
df
```

Problem_03

```python
#Handling Missing Data: From Age,Salary Columns
#Counting missing values
print("Count of Missing Values")
df[["Age","Salary"]].isnull().sum()
```

```python
#Filling missing age with mean age
df["Age"].fillna(df["Age"].mean(),inplace=True)
df
```

```python
#Dropping rows where Salary is missing
df = df.dropna(subset="Salary")
df
```

Problem_04

```python
#Applying statistical Functions:
#Calculating mean,max for each subject
print("Mean for each subject")
print(df[["Marks1","Marks2","Marks3"]].mean())
print("Max for each subject")
print(df[["Marks1","Marks2","Marks3"]].max())
```

```python
#Total marks per student
df["Total Marks"] = df.loc[:,["Marks1","Marks2","Marks3"]].sum(axis = 1)
df
```

```python
print("Standard deviation of total marks:")
print(df["Total Marks"].std())
```
