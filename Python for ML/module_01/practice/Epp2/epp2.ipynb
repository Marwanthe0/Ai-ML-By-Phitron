```python
with open("names.txt","r") as f:
    content = f.read()
    content = list(content.split())
    content = list(map(lambda x:(x.replace(" ","")).capitalize(),content))
    content = sorted(list(set(content)))
    print(content)
    with open("clean_names.txt","w") as cf:
        cf.write("\n".join(content))
```
