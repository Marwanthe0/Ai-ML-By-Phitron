```python
a_friends = {"Rahim", "Karim", "Sakib", "Jamal"}
b_friends = {"Sakib", "Jamal", "Rafiq", "Nadim"}
mutual_firends = a_friends.intersection(b_friends)
print("mutual firends: ",mutual_firends)
unique_to_a = a_friends.difference(mutual_firends)
unique_to_b = b_friends.difference(mutual_firends)
print("unique to a: ",unique_to_a)
print("unique to b: ",unique_to_b)
print("Total unique friends: ",len(a_friends.union(b_friends)))
```
