```python
students = [("Rafi", 89), ("Sumi", 95), ("Hasan", 90), ("Nila", 75), ("Anik", 98)]
dct = dict(students)
dct2 = {y: x for x, y in dct.items()}
keys = sorted(list(dct2.keys()), reverse=True)
ans = {dct2[x]: x for x in keys}
ans = [x for x in ans.items()]
ans = ans[:3]
ans = dict(ans)
print("Top 3 Students are:")
for x, y in ans.items():
    print(f"{x}  -  {y}")

students = [("Rafi", 89), ("Sumi", 95), ("Hasan", 90), ("Nila", 75), ("Anik", 98)]
# students = list(map((lambda x: (x[1], x[0])), students))
students = [(item[1],item[0])for item in students]
students.sort(reverse=True)
students = (list(map((lambda x: (x[1], x[0])), students)))[:3]
print(students)

print("Top 3 Students are:")
for x in students:
    print(f"{x[0]}  -  {x[1]}")
```
