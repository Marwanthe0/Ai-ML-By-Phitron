```python
def ans(inp):
    inp = inp.replace(" ", "")
    dct = {}
    for letter in inp:
        dct[letter] = dct.get(letter, 0) + 1
    lst = sorted(([(x,y) for x,y in dct.items()]),key = lambda x:x[0])
    for x in lst:
        print(f"Frequency of {x[0]} is {x[1]}")


line = "jslajdflajoiwjn fjodajfo sfoajfj lsjaljd fdsjkfsda sk fajfjeke sf fka jw j d fdskaj dfsjads f dfja dfj cj dfs afs "
ans(line)
```
