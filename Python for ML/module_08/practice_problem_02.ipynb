```python
class MobilePhone:
    def __init__(self, model, battery=100):
        self.model = model
        self.__battery = battery

    def use(self,percent):
        self.__battery = max(0, self.__battery - percent)

    def charge(self,percent=100):
        self.__battery = min(100, self.__battery + percent)
    def get_battery(self):
        return self.__battery
    
phone =  MobilePhone("iqoo Neo 10",80)
#phone.battery can't be accesed
# print(phone.battery)
phone.use(10)
phone.use(50)
phone.use(23)
phone.charge(2)
phone.charge(2)
phone.charge(2)
print(phone.get_battery())
```
