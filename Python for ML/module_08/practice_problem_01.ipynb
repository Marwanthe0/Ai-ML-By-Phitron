```python
class Phone():
    def __init__(self,model,battery = 100):
        self.model = model
        self.battery = battery

class SmartPhone(Phone):
    def __init__(self,model,operating_system,battery = 100):
        super().__init__(model,battery)
        self.operating_system = operating_system

class GamingPhone(SmartPhone):
    def __init__(self,model,operating_system,cooling_system,battery = 100):
        super().__init__(model,operating_system,battery)
        self.cooling_system = cooling_system
    def start_game(self,name):
        print(f"playing{self.name} on {self.model}")

phone1 = Phone("Nothing cmf 2 pro")
phone2 = SmartPhone("Redmi note 14 pro","Android")
phone3 = GamingPhone("Asus Rog phone 7","Android","Liquid Cooling Technology")
print(f"Frist phone is {phone1.model} and has {phone1.battery}% of Battery.It is a {phone1.__class__.__name__}")
print(f"Second phone is {phone2.model} Runs on {phone2.operating_system} and has {phone1.battery}% of Battery. It is a {phone2.__class__.__name__}")
print(f"Third Gaming phone is {phone3.model} Runs on {phone2.operating_system}, Uses {phone3.cooling_system} and has {phone1.battery}% of Battery.It is a {phone3.__class__.__name__}.")
```
