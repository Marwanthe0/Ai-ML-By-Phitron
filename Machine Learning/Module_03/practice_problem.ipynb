Problem 1 — Scaling head-to-head

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

 a) Compute the Standardization for both features by hand on the first three values only.

```python
weight = [58, 62, 65, 66, 190]
height = [150, 160, 170, 175, 180]
```

```python
w_array = np.array(weight)
h_array = np.array(height)
```

```python
df = pd.DataFrame({"weight":weight,"height":height})
df
```

```python
mean = df.mean()
std = df.std()
mean,std
```

```python
df[["Z_scale_weight", "Z_Scale_height"]] = (df - mean) / std
df
```

 b) Compute Min–Max Scaling to [0, 1] for all values.

```python
main_keys = ["weight","height"];
mn = df[main_keys].min()
mx = df[main_keys].max()
mn,mx
df[["mm_Scale_weight","mm_Scale_height"]] = (df[main_keys] - mn)/(mx - mn)
df
```

 c) Compute Robust Scaling for Weight_kg using median and IQR.


```python
median = df["weight"].median()
q1 = df["weight"].quantile(.25)
q3 = df["weight"].quantile(.75)
iqr = q3 - q1
df["Robust_Scale_weight"] = (df["weight"] - median)/iqr
df
```

**Problem 4 — Vectors, dot product, norms**

Given vectors a = [3, −1, 2] and b = [4, 0, −2].

**Tasks:**

 a) Compute the dot product a·b by hand.


```python
a = [3, -1, 2]
b = [4, 0, -2]
a = np.array(a)
b = np.array(b)
```

```python
dot_ab = (a*b).sum()
print(dot_ab)
print(a@b)
```

 b) Compute L1 and L2 norms of a.

```python
l1_norm_hand = np.sqrt(np.sum(a**2))
print(l1_norm_hand)
l2_norm_hand = np.sum(np.abs(a))
print(l2_norm_hand)
```

```python
l2_norm = np.linalg.norm(a)
l1_norm = np.linalg.norm(a,ord=1)
print(l1_norm)
print(l2_norm)
```

Problem 5 — Euclidean vs Manhattan distance in practice

```python
points = [(2,3),(5,7),(2,10)]
p1 = [x[0] for x in points]
p2 = [x[1] for x in points]
arr1 = np.array(p1)
arr2 = np.array(p2)
arr1,arr2
```

 a) Compute Euclidean and Manhattan distances for all pairs by hand.

```python
eu_d = np.sqrt(np.sum((arr1 - arr2)**2))
man_d = np.sum(np.abs(arr1 - arr2))
print(eu_d)
print(man_d)
```
